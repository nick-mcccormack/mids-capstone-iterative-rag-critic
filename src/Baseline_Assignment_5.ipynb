## 2. Building the Components of our RAG System

Let us introduce and test the base components of our RAG system. We will largely use the Hugging Face and LangChain libraries.



### 2.1 The Embedding Model

We will need to represent text (pieces) as vectors. For this, we will use the [sentence_transformer](https://sbert.net/docs/sentence_transformer/pretrained_models.html) architecture.



**NOTE:** The embedding models you can use are: 'all-mpnet-base-v2', 'all-MiniLM-L12-v2', 'multi-qa-mpnet-base-dot-v1', 'all-distilroberta-v1', and 'multi-qa-distilbert-cos-v1 '

```python
text = "This is a test document."
query_result = base_embeddings.embed_query(text)
print(f'Embedding dimension: {len(query_result)}')
print(query_result)

doc_result = base_embeddings.embed_documents(["Germany won the World Cup 4 times.", "This is a training document."])
len(doc_result)
```

### 2.2. Loading and Chunking Texts

We first need to load the documents. Here is an example:

```python
loader = WebBaseLoader(
    web_paths=("https://lilianweng.github.io/posts/2023-06-23-agent/",),
    bs_kwargs=dict(
        parse_only=bs4.SoupStrainer(
            class_=("post-content", "post-title", "post-header")
        )
    ),
)

documents = loader.load()
```

We will need to split the  text in chunks that are 'suitable' as retrieval units. Let's for starters define a chunk size of 128 and have no overlap between the chunks:  


```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=128, chunk_overlap=0)
splits = text_splitter.split_documents(documents)
print('Number of splits/chunks: ', str(len(splits)))
```

Ok, so it looks like we have now many splits (chunks) from one document. Here is how you can get the content:

```python
splits[39].page_content
```

Perfect. Now we have the splits and embeddings. Next, the embeddings need to be stored in a vector db.

### 2.3 Storing the Embeddings of Chunks in Vectorstores

After loading and chunking the data, we need to save the vector representations of the chunks in a vectorstore. We will use Qdrant here for simplicity. We load the splits (structured chunks) and the embeddings:

```python

vectorstore = Qdrant.from_documents(splits,
    base_embeddings,
    location=":memory:",  # Local mode with in-memory storage only
    collection_name="test",
)
retriever = vectorstore.as_retriever()
```

The nice thing is that the vector store also does the similarity searches for us:

```python
query = "What is Chain of Thought doing?"
docs = vectorstore.similarity_search_by_vector(base_embeddings.embed_query(query)) # will rank the splits
```

```python
docs
```

Looks good! We have an ordered list of documents that seem to relate to the question. That is what we need.

### 2.5 Testing the LLM in a LangChain Chain

Chains will be defined and discussed in Week 10. In short, they are convenient programmatic ways to deal with 'chains' of actions that involve LLMs. For example, a list of events like 'here is a city name. Plug that city name into prompt template, then generate a story about that city. Lastly, format the model output as a string' can be easily handled by LangChain's Chain framework. In this case, the Chain would consist of the prompt template, the LLM, and the String Formatter. The parameter (the city in this case) will be provided at run time by invocation of the Chain. Let's test that.

```python
mistral_llm_lc = HuggingFacePipeline(pipeline=mistral_pipe)
```

### 2.6 Setting Up a Simple RAG Chain

For RAG, we will follow the same approach. Except... you will **later** need to change the chain to include the retrieval step.

We first do a simple test: create a RAG template that takes a question and a pre-defined context as input, and generates the answer based on the provided context:

```python
rag_template = """[INST] Answer the question based only on the following context:
{context}

Question: {question}
[/INST]
"""
rag_prompt_template = ChatPromptTemplate.from_template(rag_template)

base_rag_chain =(
    {"context": RunnablePassthrough(),
     "question": RunnablePassthrough()}
    | rag_prompt_template
    | mistral_llm_lc
    | output_parser
)

predefined_context = "Germany has won the World Cup 4 times."
question = "How many times did Germany win the world cup?"

resp = base_rag_chain.invoke({'context': predefined_context,
                           'question': question})
print(resp)
```

That's great. But of course, the context needs to be created in an earlier retrieval step. More precisely, the documents will be first retrieved as a list, and then they will need to be formatted into one string to pass to the LLM in the context window.

Here is a simple formatting function that can be hooked into the chain, which combines a list of chunks into one string:



```python
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)
```

So how could we build a simple chain? Let's first just get the retrieval done and the formatted retrieved data and the question inserted into the prompt template:

```python
rag_template = """Here is a context:\n{context} \n\nand here is a question: \n{question}"""

rag_prompt = ChatPromptTemplate.from_template(rag_template)

rag_chain = (
    {"context": retriever | format_docs,
     "question": RunnablePassthrough()}
    | rag_prompt


)
```

```python
output = rag_chain.invoke('What is Chain of Thought?')
```

Ok... with some formatting... this looks good:

```python
print(output.messages[0].content)
```

Let's complete the RAG Chain:

```python
output_parser = StrOutputParser()

rag_template = """[INST]Please answer the question below only based on the context information provided.\n\nHere is a context:\n{context} \n\nHere is a question: \n{question}.[/INST]"""
rag_prompt = ChatPromptTemplate.from_template(rag_template)

rag_chain = (
    {"context": retriever | format_docs,
     "question": RunnablePassthrough()}
    | rag_prompt
    | mistral_llm_lc

)

```

```python
rag_chain.invoke('What is Chain of Thought?')
```

```python
## Test

# Enhanced RAG Pipeline using your existing components
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

class EnhancedRAGPipeline:
    def __init__(self, vectorstore, retriever, llm_mistral, llm_cohere):
        self.vectorstore = vectorstore
        self.retriever = retriever
        self.llm_mistral = llm_mistral
        self.llm_cohere = llm_cohere
        self.output_parser = StrOutputParser()

        # Define user-specific prompt templates
        self.engineering_template = """[INST] You are an AI assistant helping engineering researchers. Provide detailed, technical responses with specifics about implementations, methodologies, and technical considerations.

Based on the following context, answer the question with technical depth and precision:

Context:
{context}

Question: {question}

Provide a comprehensive technical answer that includes:
- Detailed explanations of concepts
- Technical implementation details when relevant
- References to specific methodologies or frameworks
- Potential limitations or considerations
[/INST]"""

        self.marketing_template = """[INST] You are an AI assistant helping marketing and business teams understand technical concepts. Provide clear, accessible explanations that focus on business value and practical applications.

Based on the following context, answer the question in a way that's accessible to non-technical stakeholders:

Context:
{context}

Question: {question}

Provide a clear, business-focused answer that includes:
- Simple explanations of technical concepts
- Business value and practical applications
- Key benefits and use cases
- Accessible language without heavy technical jargon
[/INST]"""

        # Create prompt templates
        self.engineering_prompt = ChatPromptTemplate.from_template(self.engineering_template)
        self.marketing_prompt = ChatPromptTemplate.from_template(self.marketing_template)

    def format_docs(self, docs):
        """Format retrieved documents with metadata"""
        formatted_docs = []
        for i, doc in enumerate(docs):
            source = doc.metadata.get('doc_source', 'Unknown')
            doc_num = doc.metadata.get('doc_num', 'Unknown')
            formatted_docs.append(f"[Document {i+1} - Source: {source}, Doc #{doc_num}]\n{doc.page_content}")
        return "\n\n".join(formatted_docs)

    def get_retriever_with_params(self, k=5, search_type="similarity"):
        """Get retriever with specific parameters"""
        return self.vectorstore.as_retriever(
            search_type=search_type,
            search_kwargs={"k": k}
        )

    def create_rag_chain(self, user_type="engineering", llm_choice="mistral", k=5):
        """Create RAG chain for specific configuration"""
        # Select prompt template
        prompt_template = self.engineering_prompt if user_type == "engineering" else self.marketing_prompt

        # Select LLM
        llm = self.llm_mistral if llm_choice == "mistral" else self.llm_cohere

        # Get retriever with specific k
        retriever = self.get_retriever_with_params(k=k)

        # Create the RAG chain
        rag_chain = (
            {"context": retriever | self.format_docs, "question": RunnablePassthrough()}
            | prompt_template
            | llm
            | self.output_parser
        )

        return rag_chain, retriever

    def generate_response(self, question, user_type="engineering", llm_choice="mistral", k=5):
        """Generate response and return detailed results"""
        # Create RAG chain
        rag_chain, retriever = self.create_rag_chain(user_type, llm_choice, k)

        # Get retrieved documents for analysis
        retrieved_docs = retriever.invoke(question)
        context = self.format_docs(retrieved_docs)

        # Generate response
        try:
            response = rag_chain.invoke(question)
        except Exception as e:
            response = f"Error generating response: {str(e)}"

        return {
            'question': question,
            'user_type': user_type,
            'llm_choice': llm_choice,
            'k': k,
            'context': context,
            'response': response,
            'retrieved_docs': retrieved_docs,
            'doc_sources': [doc.metadata.get('doc_source', 'Unknown') for doc in retrieved_docs]
        }

# Initialize the enhanced RAG pipeline with your components
rag_pipeline = EnhancedRAGPipeline(
    vectorstore=qdrant_vectorstore,
    retriever=retriever,
    llm_mistral=mistral_llm_lc,
    llm_cohere=cohere_chat_model
)

print("✅ Enhanced RAG Pipeline initialized with your components")
```

```python
import time
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from bert_score import score as bert_score
from rouge_score import rouge_scorer
import nltk
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from sentence_transformers import SentenceTransformer
import warnings
import logging
warnings.filterwarnings('ignore')

# Set up logging for debugging
logging.basicConfig(level=logging.INFO)

# Download required NLTK data
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)

def clean_contexts(context_string):
    """Convert context string to list of contexts"""
    if isinstance(context_string, list):
        return [str(c).strip() for c in context_string if c]

    contexts = []
    if '[Document' in context_string:
        parts = context_string.split('[Document')
        for part in parts[1:]:
            lines = part.split('\n')
            content = []
            for line in lines:
                if line.strip() and not any(skip in line for skip in ['Source:', 'Doc #', ']']):
                    content.append(line.strip())
            if content:
                contexts.append(' '.join(content))
    else:
        contexts = [context_string.strip()]

    return contexts[:5]  # Max 5 contexts

def compute_bert_score(generated, reference):
    """Compute BERTScore between generated and reference text"""
    try:
        P, R, F1 = bert_score([generated], [reference], lang="en",
                             model_type="microsoft/deberta-v3-base",
                             device='cpu', rescale_with_baseline=True)
        return {
            'bert_precision': float(P[0]),
            'bert_recall': float(R[0]),
            'bert_f1': float(F1[0])
        }
    except Exception as e:
        logging.error(f"BERTScore error: {e}")
        return {'bert_precision': None, 'bert_recall': None, 'bert_f1': None}

def compute_rouge_scores(generated, reference):
    """Compute ROUGE scores"""
    try:
        scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
        scores = scorer.score(reference, generated)
        return {
            # Fixed: Use consistent naming
            'rouge1': scores['rouge1'].fmeasure,
            'rouge2': scores['rouge2'].fmeasure,
            'rougeL': scores['rougeL'].fmeasure
        }
    except Exception as e:
        logging.error(f"ROUGE error: {e}")
        return {'rouge1': None, 'rouge2': None, 'rougeL': None}

def compute_bleu_score(generated, reference):
    """Compute BLEU score"""
    try:
        from nltk.tokenize import word_tokenize
        ref_tokens = word_tokenize(reference.lower())
        gen_tokens = word_tokenize(generated.lower())

        if not ref_tokens or not gen_tokens:
            return 0.0

        smoothing = SmoothingFunction()
        return sentence_bleu([ref_tokens], gen_tokens, smoothing_function=smoothing.method1)
    except Exception as e:
        logging.error(f"BLEU error: {e}")
        return None

def compute_semantic_similarity(text1, text2, embeddings_model):
    """Compute semantic similarity using embeddings"""
    try:
        # Handle different embedding model types
        if hasattr(embeddings_model, 'encode'):
            # SentenceTransformer
            emb1 = embeddings_model.encode(text1, convert_to_numpy=True)
            emb2 = embeddings_model.encode(text2, convert_to_numpy=True)
        elif hasattr(embeddings_model, 'embed_documents'):
            # HuggingFaceEmbeddings
            emb1 = np.array(embeddings_model.embed_documents([text1])[0])
            emb2 = np.array(embeddings_model.embed_documents([text2])[0])
        else:
            logging.error(f"Unknown embeddings model type: {type(embeddings_model)}")
            return None

        similarity = cosine_similarity([emb1], [emb2])[0][0]
        return float(similarity)
    except Exception as e:
        logging.error(f"Semantic similarity error: {e}")
        return None

def compute_keyword_overlap(generated, reference):
    """Compute keyword overlap ratio"""
    try:
        gen_words = set([w.lower() for w in generated.split() if len(w) > 3 and w.isalnum()])
        ref_words = set([w.lower() for w in reference.split() if len(w) > 3 and w.isalnum()])

        if not ref_words:
            return 0.0

        return len(gen_words & ref_words) / len(ref_words)
    except Exception as e:
        logging.error(f"Keyword overlap error: {e}")
        return None

def compute_context_utilization(answer, context, embeddings_model):
    """Measure how well the answer uses the provided context"""
    try:
        context_sim = compute_semantic_similarity(answer, context, embeddings_model)

        context_words = set([w.lower() for w in context.split() if len(w) > 4 and w.isalnum()])
        answer_words = set([w.lower() for w in answer.split() if len(w) > 4 and w.isalnum()])

        if answer_words:
            word_usage = len(answer_words & context_words) / len(answer_words)
        else:
            word_usage = 0.0

        return (context_sim + word_usage) / 2 if context_sim is not None else word_usage
    except Exception as e:
        logging.error(f"Context utilization error: {e}")
        return None

def compute_hallucination_score(answer, context):
    """Estimate hallucination by checking unsupported claims"""
    try:
        context_lower = context.lower()
        sentences = answer.split('.')
        unsupported_count = 0
        total_claims = 0

        for sentence in sentences:
            sentence = sentence.strip()
            if not sentence:
                continue

            # Check if sentence contains factual claims
            has_claim = any([
                any(char.isdigit() for char in sentence),
                any(indicator in sentence.lower() for indicator in [
                    'percent', '%', 'million', 'billion', 'first', 'largest', 'only'
                ])
            ])

            if has_claim:
                total_claims += 1
                key_terms = [w.lower() for w in sentence.split() if len(w) > 4 and w.isalnum()]

                if key_terms:
                    support_score = sum(1 for term in key_terms if term in context_lower) / len(key_terms)
                    if support_score < 0.3:  # Less than 30% of terms found in context
                        unsupported_count += 1

        if total_claims == 0:
            return 0.0  # No claims to evaluate

        return unsupported_count / total_claims
    except Exception as e:
        logging.error(f"Hallucination score error: {e}")
        return None

def evaluate_rag_response(question, generated_answer, gold_answer, context, embeddings_model,
                         use_bert_score=True, use_rouge=True):
    """Comprehensive evaluation of a RAG response"""
    logging.info("Evaluating response...")
    start_time = time.time()

    results = {}

    # Basic length metrics
    gen_length = len(generated_answer.split())
    gold_length = len(gold_answer.split())
    results['response_length'] = gen_length
    results['gold_length'] = gold_length
    results['length_ratio'] = min(gen_length, gold_length) / max(gen_length, gold_length) if gold_length > 0 else 0.0

    # Semantic similarity
    logging.info("Computing semantic similarity...")
    results['semantic_similarity'] = compute_semantic_similarity(generated_answer, gold_answer, embeddings_model)

    # Keyword overlap
    results['keyword_overlap'] = compute_keyword_overlap(generated_answer, gold_answer)

    # BERTScore (optional)
    if use_bert_score:
        logging.info("Computing BERTScore...")
        bert_scores = compute_bert_score(generated_answer, gold_answer)
        results.update(bert_scores)

    # ROUGE scores (optional)
    if use_rouge:
        logging.info("Computing ROUGE scores...")
        rouge_scores = compute_rouge_scores(generated_answer, gold_answer)
        results.update(rouge_scores)

    # BLEU score
    results['bleu_score'] = compute_bleu_score(generated_answer, gold_answer)

    # Context-based metrics
    logging.info("Computing context-based metrics...")
    results['context_utilization'] = compute_context_utilization(generated_answer, context, embeddings_model)
    results['hallucination_score'] = compute_hallucination_score(generated_answer, context)
    results['factual_consistency'] = 1 - results['hallucination_score'] if results['hallucination_score'] is not None else None

    results['evaluation_time'] = time.time() - start_time

    return results

def evaluate_batch(questions, generated_answers, gold_answers, contexts, embeddings_model,
                  use_bert_score=True, use_rouge=True):
    """Evaluate multiple responses and compute statistics"""
    all_results = []
    total = len(questions)

    for i, (q, gen, gold, ctx) in enumerate(zip(questions, generated_answers, gold_answers, contexts)):
        logging.info(f"Evaluating {i+1}/{total}...")
        result = evaluate_rag_response(
            question=q,
            generated_answer=gen,
            gold_answer=gold,
            context=ctx,
            embeddings_model=embeddings_model,
            use_bert_score=use_bert_score,
            use_rouge=use_rouge
        )
        all_results.append(result)

    # Compute aggregate statistics
    stats = compute_aggregate_stats(all_results)

    return all_results, stats

def compute_aggregate_stats(results):
    """Compute mean, std, min, max for all metrics across results"""
    stats = {}

    if results:
        metric_names = list(results[0].keys())
        for metric in metric_names:
            values = [r[metric] for r in results if r.get(metric) is not None]
            if values and metric != 'evaluation_time':
                stats[metric] = {
                    'mean': np.mean(values),
                    'std': np.std(values),
                    'min': min(values),
                    'max': max(values),
                    'count': len(values)
                }

    return stats

def print_evaluation_results(results, title="Evaluation Results"):
    """Pretty print evaluation results"""
    print(f"\n{title}")
    print("=" * 60)

    metric_groups = {
        "Similarity Metrics": ['semantic_similarity', 'bert_f1', 'rouge1', 'rouge2', 'rougeL', 'bleu_score', 'keyword_overlap'],
        "Context Metrics": ['context_utilization', 'factual_consistency', 'hallucination_score'],
        "Length Metrics": ['response_length', 'gold_length', 'length_ratio']
    }

    for group_name, metrics in metric_groups.items():
        print(f"\n{group_name}:")
        print("-" * 40)
        for metric in metrics:
            if metric in results:
                value = results[metric]
                if value is not None:
                    if isinstance(value, float):
                        print(f"  {metric}: {value:.3f}")
                    else:
                        print(f"  {metric}: {value}")
                else:
                    print(f"  {metric}: N/A")

    if 'evaluation_time' in results:
        print(f"\nTotal evaluation time: {results['evaluation_time']:.2f}s")

def print_aggregate_stats(stats, title="Aggregate Statistics"):
    """Pretty print aggregate statistics"""
    print(f"\n{title}")
    print("=" * 60)
    print(f"{'Metric':<25} {'Mean':>8} {'Std':>8} {'Min':>8} {'Max':>8}")
    print("-" * 60)

    for metric, values in stats.items():
        print(f"{metric:<25} {values['mean']:>8.3f} {values['std']:>8.3f} "
              f"{values['min']:>8.3f} {values['max']:>8.3f}")
```

```python
import pandas as pd
import numpy as np
import time
from datetime import datetime

def evaluate_all_questions_comprehensive(validation_questions_answers,
                                       eng_rag_function,
                                       marketing_rag_function,
                                       embeddings_model,
                                       use_bert_score=False,
                                       use_rouge=True):
    """
    Evaluate all questions with both models and return comprehensive DataFrame

    Args:
        validation_questions_answers: Your validation dict
        eng_rag_function: Function that takes question and returns {'response': str, 'context': str}
        marketing_rag_function: Function that takes question and returns {'response': str, 'context': str}
        embeddings_model: Your embedding model
        use_bert_score: Whether to include BERT scores (slower)
        use_rouge: Whether to include ROUGE scores

    Returns:
        DataFrame with all metrics for all questions
    """

    all_results = []
    session_id = datetime.now().strftime("%Y%m%d_%H%M%S")
    total_questions = len(validation_questions_answers)

    print(f"Starting comprehensive evaluation of {total_questions} questions")
    print("=" * 80)

    for q_id, q_data in validation_questions_answers.items():
        question = q_data['question']
        gold_eng = q_data['gold_answer_research']
        gold_marketing = q_data['gold_answer_marketing']

        print(f"\nProcessing Question {q_id}: {question[:100]}...")
        print(f"Gold Engineering Answer Length: {len(gold_eng.split())} words")
        print(f"Gold Marketing Answer Length: {len(gold_marketing.split())} words")

        # Get RAG responses
        print("Getting Engineering response...")
        eng_result = eng_rag_function(question)

        print("Getting Marketing response...")
        marketing_result = marketing_rag_function(question)

        print("\n" + "=" * 60)
        print("TESTING ENGINEERING RESPONSE:")
        print("-" * 40)

        # Evaluate Engineering model
        eng_eval = evaluate_rag_response(
            question=question,
            generated_answer=eng_result['response'],
            gold_answer=gold_eng,
            context=eng_result['context'],
            embeddings_model=embeddings_model,
            use_bert_score=use_bert_score,
            use_rouge=use_rouge
        )

        print("\nENGINEERING EVALUATION RESULTS:")
        key_metrics = [
            'semantic_similarity', 'keyword_overlap', 'length_ratio',
            'factual_consistency', 'hallucination_score'
        ]

        for metric in key_metrics:
            if metric in eng_eval:
                value = eng_eval[metric]
                if value is not None:
                    print(f"  {metric}: {value:.3f}")
                else:
                    print(f"  {metric}: N/A")

        print(f"\n  Response length: {eng_eval['response_length']} words")
        print(f"  Gold length: {eng_eval['gold_length']} words")

        print("\n" + "=" * 60)
        print("TESTING MARKETING RESPONSE:")
        print("-" * 40)

        # Evaluate Marketing model
        marketing_eval = evaluate_rag_response(
            question=question,
            generated_answer=marketing_result['response'],
            gold_answer=gold_marketing,
            context=marketing_result['context'],
            embeddings_model=embeddings_model,
            use_bert_score=use_bert_score,
            use_rouge=use_rouge
        )

        print("\nMARKETING EVALUATION RESULTS:")
        for metric in key_metrics:
            if metric in marketing_eval:
                value = marketing_eval[metric]
                if value is not None:
                    print(f"  {metric}: {value:.3f}")
                else:
                    print(f"  {metric}: N/A")

        print(f"\n  Response length: {marketing_eval['response_length']} words")
        print(f"  Gold length: {marketing_eval['gold_length']} words")

        print("\n" + "=" * 60)
        print("COMPARISON:")
        print("-" * 40)

        # Compare key metrics
        comparison_metrics = ['semantic_similarity', 'factual_consistency', 'keyword_overlap']
        print(f"{'Metric':<20} {'Engineering':>12} {'Marketing':>12} {'Better':>12}")
        print("-" * 56)

        for metric in comparison_metrics:
            eng_val = eng_eval.get(metric)
            mark_val = marketing_eval.get(metric)

            if eng_val is not None and mark_val is not None:
                better = "Engineering" if eng_val > mark_val else "Marketing"
                print(f"{metric:<20} {eng_val:>12.3f} {mark_val:>12.3f} {better:>12}")
            else:
                eng_str = f"{eng_val:.3f}" if eng_val is not None else "N/A"
                mark_str = f"{mark_val:.3f}" if mark_val is not None else "N/A"
                print(f"{metric:<20} {eng_str:>12} {mark_str:>12} {'N/A':>12}")

        print("\n" + "=" * 60)
        print("DETAILED ROUGE SCORES:")
        print("-" * 40)

        rouge_metrics = ['rouge1', 'rouge2', 'rougeL']
        print(f"{'ROUGE Metric':<15} {'Engineering':>12} {'Marketing':>12}")
        print("-" * 39)

        for metric in rouge_metrics:
            eng_val = eng_eval.get(metric)
            mark_val = marketing_eval.get(metric)

            eng_str = f"{eng_val:.3f}" if eng_val is not None else "N/A"
            mark_str = f"{mark_val:.3f}" if mark_val is not None else "N/A"
            print(f"{metric:<15} {eng_str:>12} {mark_str:>12}")

        print(f"\nEvaluation complete for Question {q_id}")
        print(f"Engineering evaluation time: {eng_eval['evaluation_time']:.2f}s")
        print(f"Marketing evaluation time: {marketing_eval['evaluation_time']:.2f}s")

        # Create engineering row
        eng_row = {
            'session_id': session_id,
            'timestamp': datetime.now().isoformat(),
            'question_id': q_id,
            'model_type': 'engineering',
            'question_text': question,
            'generated_answer': eng_result['response'],
            'gold_answer': gold_eng,
            'context': eng_result['context'],
            'question_length_words': len(question.split()),
            'generated_length_words': len(eng_result['response'].split()),
            'gold_length_words': len(gold_eng.split()),
            'context_length_words': len(eng_result['context'].split()),
            'question_length_chars': len(question),
            'generated_length_chars': len(eng_result['response']),
            'gold_length_chars': len(gold_eng),
            'context_length_chars': len(eng_result['context'])
        }

        # Add all engineering metrics
        for metric, value in eng_eval.items():
            eng_row[f'metric_{metric}'] = value

        # Create marketing row
        marketing_row = {
            'session_id': session_id,
            'timestamp': datetime.now().isoformat(),
            'question_id': q_id,
            'model_type': 'marketing',
            'question_text': question,
            'generated_answer': marketing_result['response'],
            'gold_answer': gold_marketing,
            'context': marketing_result['context'],
            'question_length_words': len(question.split()),
            'generated_length_words': len(marketing_result['response'].split()),
            'gold_length_words': len(gold_marketing.split()),
            'context_length_words': len(marketing_result['context'].split()),
            'question_length_chars': len(question),
            'generated_length_chars': len(marketing_result['response']),
            'gold_length_chars': len(gold_marketing),
            'context_length_chars': len(marketing_result['context'])
        }

        # Add all marketing metrics
        for metric, value in marketing_eval.items():
            marketing_row[f'metric_{metric}'] = value

        all_results.append(eng_row)
        all_results.append(marketing_row)

        print(f"Progress: {len(all_results)//2}/{total_questions} questions completed")
        print("=" * 80)

    # Create DataFrame
    df = pd.DataFrame(all_results)

    print(f"Evaluation complete! Created DataFrame with {len(df)} rows")
    print(f"Columns: {len(df.columns)}")
    print(f"Engineering evaluations: {len(df[df['model_type'] == 'engineering'])}")
    print(f"Marketing evaluations: {len(df[df['model_type'] == 'marketing'])}")

    return df

def save_evaluation_results(df, filename_prefix="rag_evaluation"):
    """
    Save evaluation DataFrame to multiple formats
    """
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

    # Save as CSV
    csv_filename = f"{filename_prefix}_{timestamp}.csv"
    df.to_csv(csv_filename, index=False)
    print(f"Saved CSV: {csv_filename}")

    # Save as Excel with multiple sheets
    excel_filename = f"{filename_prefix}_{timestamp}.xlsx"
    with pd.ExcelWriter(excel_filename) as writer:
        # All data
        df.to_excel(writer, sheet_name='All_Results', index=False)

        # Engineering only
        eng_df = df[df['model_type'] == 'engineering'].copy()
        eng_df.to_excel(writer, sheet_name='Engineering', index=False)

        # Marketing only
        marketing_df = df[df['model_type'] == 'marketing'].copy()
        marketing_df.to_excel(writer, sheet_name='Marketing', index=False)

        # Summary statistics
        summary_df = create_summary_stats(df)
        summary_df.to_excel(writer, sheet_name='Summary_Stats')

    print(f"Saved Excel: {excel_filename}")

    return csv_filename, excel_filename

def create_summary_stats(df):
    """
    Create summary statistics DataFrame
    """
    # Get metric columns
    metric_cols = [col for col in df.columns if col.startswith('metric_')]

    # Calculate stats by model type
    summary_data = []

    for model_type in ['engineering', 'marketing']:
        model_df = df[df['model_type'] == model_type]

        for metric in metric_cols:
            values = model_df[metric].dropna()
            if len(values) > 0:
                summary_data.append({
                    'model_type': model_type,
                    'metric': metric.replace('metric_', ''),
                    'count': len(values),
                    'mean': values.mean(),
                    'std': values.std(),
                    'min': values.min(),
                    'max': values.max(),
                    'median': values.median(),
                    'q25': values.quantile(0.25),
                    'q75': values.quantile(0.75)
                })

    return pd.DataFrame(summary_data)

def print_final_summary(df):
    """
    Print final summary of all results
    """
    print("\n" + "=" * 80)
    print("FINAL COMPREHENSIVE SUMMARY")
    print("=" * 80)

    # Key metrics to highlight
    key_metrics = [
        'metric_semantic_similarity',
        'metric_factual_consistency',
        'metric_keyword_overlap',
        'metric_rouge1',
        'metric_rouge2',
        'metric_rougeL',
        'metric_hallucination_score'
    ]

    print(f"{'Metric':<25} {'Engineering':>12} {'Marketing':>12} {'Better':>12}")
    print("-" * 64)

    for metric in key_metrics:
        if metric in df.columns:
            eng_mean = df[df['model_type'] == 'engineering'][metric].mean()
            mark_mean = df[df['model_type'] == 'marketing'][metric].mean()

            if not pd.isna(eng_mean) and not pd.isna(mark_mean):
                better = "Engineering" if eng_mean > mark_mean else "Marketing"
                # For hallucination, lower is better
                if 'hallucination' in metric:
                    better = "Engineering" if eng_mean < mark_mean else "Marketing"

                print(f"{metric.replace('metric_', ''):<25} {eng_mean:>12.3f} {mark_mean:>12.3f} {better:>12}")

    print("\nResponse Length Analysis:")
    print("-" * 40)
    eng_lengths = df[df['model_type'] == 'engineering']['generated_length_words']
    mark_lengths = df[df['model_type'] == 'marketing']['generated_length_words']

    print(f"Engineering avg length: {eng_lengths.mean():.1f} words")
    print(f"Marketing avg length:   {mark_lengths.mean():.1f} words")

    print(f"\nEvaluation Time Analysis:")
    print("-" * 40)
    if 'metric_evaluation_time' in df.columns:
        eng_time = df[df['model_type'] == 'engineering']['metric_evaluation_time'].sum()
        mark_time = df[df['model_type'] == 'marketing']['metric_evaluation_time'].sum()
        total_time = eng_time + mark_time

        print(f"Total evaluation time: {total_time:.1f} seconds")
        print(f"Average per question:  {total_time / (len(df) / 2):.1f} seconds")

    print(f"\nDataFrame Summary:")
    print("-" * 40)
    print(f"Total rows: {len(df)}")
    print(f"Total questions evaluated: {len(df) // 2}")
    print(f"Total columns: {len(df.columns)}")

    metric_cols = [col for col in df.columns if col.startswith('metric_')]
    print(f"Available metrics: {len(metric_cols)}")

    print("\nMetric columns:")
    for i, col in enumerate(metric_cols):
        print(f"  {i+1:2d}. {col.replace('metric_', '')}")

# Usage with your existing EnhancedRAGPipeline
def run_comprehensive_evaluation():
    """
    Main function to run comprehensive evaluation using your EnhancedRAGPipeline
    Assumes you have already defined:
    - validation_questions_answers
    - base_embeddings
    - rag_pipeline (your EnhancedRAGPipeline instance)
    """

    def eng_rag_function(question):
        """Engineering RAG function using your pipeline"""
        result = rag_pipeline.generate_response(
            question=question,
            user_type="engineering",
            llm_choice="mistral",
            k=5
        )
        return {
            'response': result['response'],
            'context': result['context']
        }

    def marketing_rag_function(question):
        """Marketing RAG function using your pipeline"""
        result = rag_pipeline.generate_response(
            question=question,
            user_type="marketing",
            llm_choice="mistral",
            k=5
        )
        return {
            'response': result['response'],
            'context': result['context']
        }

    print("Starting comprehensive RAG evaluation...")

    # Run evaluation
    results_df = evaluate_all_questions_comprehensive(
        validation_questions_answers=validation_questions_answers,
        eng_rag_function=eng_rag_function,
        marketing_rag_function=marketing_rag_function,
        embeddings_model=base_embeddings,
        use_bert_score=False,  # Set to True if you want BERT scores (slower)
        use_rouge=True
    )

    # Print final summary
    print_final_summary(results_df)

    # Save results
    csv_file, excel_file = save_evaluation_results(results_df, filename_prefix="comprehensive_rag_eval")

    print(f"\nFiles saved:")
    print(f"CSV: {csv_file}")
    print(f"Excel: {excel_file}")

    return results_df

# Alternative: Run with different configurations
def run_comprehensive_evaluation_with_configs():
    """
    Run evaluation with multiple configurations of your RAG pipeline
    """

    configs = [
        {"user_type": "engineering", "llm_choice": "mistral", "k": 5},
        {"user_type": "marketing", "llm_choice": "mistral", "k": 5},
        {"user_type": "engineering", "llm_choice": "cohere", "k": 5},
        {"user_type": "marketing", "llm_choice": "cohere", "k": 5}
    ]

    all_results = []

    for config in configs:
        print(f"\nRunning evaluation for: {config}")

        def rag_function(question):
            result = rag_pipeline.generate_response(
                question=question,
                user_type=config["user_type"],
                llm_choice=config["llm_choice"],
                k=config["k"]
            )
            return {
                'response': result['response'],
                'context': result['context']
            }

        # Run for just engineering or marketing based on config
        if config["user_type"] == "engineering":
            config_results = evaluate_all_questions_comprehensive(
                validation_questions_answers=validation_questions_answers,
                eng_rag_function=rag_function,
                marketing_rag_function=lambda q: {"response": "", "context": ""},  # Dummy
                embeddings_model=base_embeddings,
                use_bert_score=False,
                use_rouge=True
            )
            # Filter to only engineering results
            config_results = config_results[config_results['model_type'] == 'engineering']
        else:
            config_results = evaluate_all_questions_comprehensive(
                validation_questions_answers=validation_questions_answers,
                eng_rag_function=lambda q: {"response": "", "context": ""},  # Dummy
                marketing_rag_function=rag_function,
                embeddings_model=base_embeddings,
                use_bert_score=False,
                use_rouge=True
            )
            # Filter to only marketing results
            config_results = config_results[config_results['model_type'] == 'marketing']

        # Add config info
        for key, value in config.items():
            config_results[f'config_{key}'] = value

        all_results.append(config_results)

    # Combine all results
    final_df = pd.concat(all_results, ignore_index=True)

    # Save results
    csv_file, excel_file = save_evaluation_results(final_df, filename_prefix="multi_config_rag_eval")

    return final_df

# Simple function to run just your standard engineering vs marketing comparison
def run_standard_evaluation():
    """
    Run the standard engineering vs marketing evaluation using your RAG pipeline
    """

    def eng_rag_function(question):
        result = rag_pipeline.generate_response(
            question=question,
            user_type="engineering",
            llm_choice="mistral",
            k=5
        )
        return {'response': result['response'], 'context': result['context']}

    def marketing_rag_function(question):
        result = rag_pipeline.generate_response(
            question=question,
            user_type="marketing",
            llm_choice="mistral",
            k=5
        )
        return {'response': result['response'], 'context': result['context']}

    # Run evaluation
    results_df = evaluate_all_questions_comprehensive(
        validation_questions_answers=validation_questions_answers,
        eng_rag_function=eng_rag_function,
        marketing_rag_function=marketing_rag_function,
        embeddings_model=base_embeddings,
        use_bert_score=False,
        use_rouge=True
    )

    print_final_summary(results_df)
    csv_file, excel_file = save_evaluation_results(results_df)

    return results_df

# To run the standard evaluation:
# results_df = run_standard_evaluation()
```
